```python
import pandas as pd                 # Library used for working with data sets and perform data analysis.
import numpy as np                  # To perform mathematical operasion and statistics
from datetime import datetime       # Used to perform dates and time manipulation
```

```python
# Import Dataset
# Bank Churner Data
sheet_url = 'https://docs.google.com/spreadsheets/d/1Z8eV1mJZEQiNoBPnypZF1tUrrrpyxmwtPmFSnQbMweA/edit#gid=2084583872'
sheet_url_trf = sheet_url.replace('/edit#gid=', '/export?format=csv&gid=')
BCD = pd.read_csv(sheet_url_trf)
BCD.head()
```

```python
# delete column unnamed on Bank churn data
column_numbers = [x for x in range(BCD.shape[1])]
column_numbers .remove(0)
BCD_cleaned = BCD.iloc[:, column_numbers]
```

```python
# Check data type
BCD_cleaned.info()
```

```python
BCD_cleaned.head()
```

```python
# Import Dataset
# Credit Limit
sheet_url = 'https://docs.google.com/spreadsheets/d/1LUjgxIPE6dpys8ju1QeqasB1-1CAdpKVy-Z5uCYsNxU/edit#gid=1594162646'
sheet_url_trf = sheet_url.replace('/edit#gid=', '/export?format=csv&gid=')
CL = pd.read_csv(sheet_url_trf)
CL.head()
```

# Join the dataset

```python
#Merge data
df_merge = BCD_cleaned.merge(CL, how = 'inner', left_on=['CLIENTNUM'], right_on=['CLIENTNUM'])

df_merge.head()
```

```python
df_merge.info()
```

```python
df_merge.shape #Check how much column and rows
```

# Cleaning Merge Dataset

### Check Values and Typos

```python
# Check Values and Typos
print(df_merge['Attrition_Flag'].unique())
print(df_merge['Gender'].unique())
print(df_merge['Education_Level'].unique())
print(df_merge['Marital_Status'].unique())
print(df_merge['Income_Category'].unique())
print(df_merge['Card_Category'].unique())
```

```python
df_merge['Income_Category'].value_counts()
```

```python
# replace typo on income category
df_merge['Income_Category'] = df_merge['Income_Category'].str.replace('60ð¾âˆ’ 80K','$60K - $80K')

print(df_merge['Income_Category'].unique())
```

### Treating Missing Values

```python
df_merge.isnull().sum()
```

```python
# replace missing value on gender with unknown
df_merge['Gender'] = df_merge['Gender'].fillna('Unknown')

print(df_merge['Gender'].unique())
```

```python
# fillna age using mode
df_merge['Customer_Age'] = df_merge['Customer_Age'].fillna(df_merge['Customer_Age'].mode()[0])
```

```python
#recheck null values
df_merge.isnull().sum()
```

### Check Duplicate

```python
#Check the duplicate
df_merge[df_merge.duplicated()]
```

```python
# delete duplicate
df_cleaned = df_merge.copy()
df_cleaned.drop_duplicates(subset ='CLIENTNUM', keep = 'first')
```

After delete duplicate total rows is 10148 (before 10181 rows)

# EDA

```python
#Preparing a copy of df_merge for EDA
df_eda = df_cleaned.copy()
df_eda.head()

# Rounds a decimal number to 3 digits after a comma
pd.set_option('display.float_format', lambda x: '%.2f' % x)

numeric = ['Dependent_count', 'Credit_Limit_x', 'Credit_Limit_y','Total_Revolving_Bal', 'Avg_Open_To_Buy',
            'Total_Amt_Chng_Q4_Q1', 'Total_Trans_Amt', 'Total_Trans_Ct','Total_Ct_Chng_Q4_Q1','Avg_Utilization_Ratio',
            'Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_1',
            'Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_2']

desc_eda = df_eda[numeric].describe()
desc_eda.loc['kurtosis'] = df_eda[numeric].kurt() #show kurtosis statistic
desc_eda.loc['skewness'] = df_eda[numeric].skew() #show Skewness statistic
desc_eda.loc['variance'] = df_eda[numeric].var() #show variance statistic
desc_eda.round(2)
desc_eda
```

```python
df_eda[['Attrition_Flag','Gender','Education_Level','Marital_Status','Income_Category','Card_Category']].describe()
```

### Average of customer's age with income 80k-120k

```python
# Filter customer age based on income category
filtered_df_eda = df_eda[df_eda['Income_Category'].isin(['$80K - $120K'])]

# Calculate average age
average_age = filtered_df_eda['Customer_Age'].mean()

# Print the average age
print("Average of customer's age with income $80K - $120K:", average_age)
```

### Number of customers based on gender and education level

```python
#Number of Customers based on Gender
user_gender = df_eda.groupby('Gender')['CLIENTNUM'].count()
user_gender = user_gender.reset_index()
user_gender
```

```python
#Number of Customers based on Education Level
user_education_level = df_eda.groupby('Education_Level')['CLIENTNUM'].count()
user_education_level = user_education_level.reset_index()
user_education_level
```
